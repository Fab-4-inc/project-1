```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv("Crimes_-_Map.csv")
data.head()
```

```python
data.count()
```

```python
no_null_crime = data.dropna(how='any')
```

```python
no_null_crime.count()
```

```python
no_null_crime = data.groupby(" PRIMARY DESCRIPTION")
no_null_crime
```

```python
count_crime = no_null_crime[" PRIMARY DESCRIPTION"].count()
count_crime.dtype
count_crime = pd.DataFrame(no_null_crime[" PRIMARY DESCRIPTION"].count())
count_crime
```

```python
#rename columns
renamed_df = count_crime.rename(columns={" PRIMARY DESCRIPTION":"Crime Count"})
renamed_df.index.names = ['Crime Type']
renamed_df
```

```python
#bar chart
crime_chart = renamed_df.plot(kind='bar')

#X-Axis = Crime Type
crime_chart.set_xlabel("Type of Crime")
crime_chart.set_ylabel("# Crime Committed")

#Title
crime_chart.set_title("Autugst 2019 Crime Count")

#resizing
# Get current size
#fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
#print("Current size:", fig_size)
 
# Set figure width to 12 and height to 9
fig_size[0] = 7
fig_size[1] = 5
plt.rcParams["figure.figsize"] = fig_size


plt.show()
#plt.tight_layout()
```
